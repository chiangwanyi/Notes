# 梯度下降

## Gradient Descent
定义损失函数：
$cost = \frac{1}{N}\sum_{n=1}^N{\hat{y}}$

$ \frac{\partial cost}{\partial \omega} = \frac{1}{N}\sum_{n=1}^N{2x_n\omega(x_n\omega-y_n)} $

```python
import numpy as np
from typing import List
import matplotlib.pyplot as plt

# 训练数据集
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

# 初始的猜测权重
w = 1

Vector = List[float]

# 定义函数：y(hat) = w * x
def forward(x:float) -> float:
    return w * x


# 定义损失函数 Cost
def cost(xs:Vector, ys:Vector) -> float:
    cost = 0
    for x, y in(zip(xs, ys)):
        y_pred = forward(x)
        cost += (y_pred - y) ** 2
    return cost / len(xs)

# 定义梯度函数
def gradient(xs:Vector, ys:Vector) -> float:
    grad = 0
    for x, y in zip(xs, ys):
        grad += 2 * x * (x * w - y)
    return grad / len(xs)

# 学习率
a = 0.01

plt.title("Cost in each epoch") 
plt.xlabel("Epoch") 
plt.ylabel("Cost") 
plt.grid()

for epoch in range(100):
    cost_val = cost(x_data, y_data)
    grad_val = gradient(x_data, y_data)
    w -= a * grad_val
    plt.plot(epoch, cost_val, ",b")

plt.show()

print("w={}".format(w))
```
